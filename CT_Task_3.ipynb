**DATA VISUALIZATION WITH BIG DATA TOOLS**

**USE A TOOL LIKE TABLEAU OR POWER BI TO VISUALIZE INSIGHTS FROM A LARGE DATASET.**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

try:
    df = pd.read_csv('GOOG.csv')
except FileNotFoundError:
    print("Error: 'your_large_dataset.csv' not found. Please replace with your file's path.")
    exit()
```

```python
#Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Open', y='Close', hue='Volume')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.savefig('scatter_plot.png')
```

```python
#Create a bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='Open', y='Close', data=df)
plt.title('Categorical Data Bar Chart')
plt.xlabel('Category')
plt.ylabel('Value')
plt.savefig('bar_chart.png')
```

```python
# For more advanced visualizations and to integrate with Tableau/Power BI:
# 1. Save the processed data to a file type compatible with these tools (e.g., CSV, Excel).
df.to_csv('processed_data.csv', index=False)  # Save to CSV
```
